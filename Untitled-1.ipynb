```python
import requests
import pandas as pd

url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&apikey=Y5MTFA3OT2V6F3SD'
r = requests.get(url)
data = r.json()['Time Series (Daily)']
```

```python
df = pd.DataFrame.from_dict(data, orient='index', dtype= 'float32')
```

```python
df.columns = df.columns.str.replace(r'^\d+\.\s*', '', regex=True)
```

```python
df.info()
```

```python
from sqlalchemy import psycopg
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/pipelino')
```

```python
print(pd.io.sql.get_schema(df, name='staging_stock_data', con=engine))
```

```python
df.to_sql.head(n=0)(name='prod_staging_stock_data', con=engine, if_exists='replace')
```
